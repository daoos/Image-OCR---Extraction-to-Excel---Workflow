```python
# Converting and replacing png files to jpg
os.chdir(rec_path)
for file in glob.glob("*.png"):
    #print(file)
    ImageFile.LOAD_TRUNCATED_IMAGES = True
    im = Image.open(file)
    rgb_im = im.convert('RGB')
    rgb_im.save(file.replace("png", "jpg"), quality=90)
```

```python
# Image resizing for Consistency
directory = os.fsencode(inp_path)
print (datetime.datetime.utcnow())                        

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    img = cv2.imread(inp_path + filename, cv2.IMREAD_UNCHANGED)
    print (filename)
    print('Original Dimensions : ',img.shape)
    width = 1500
    height = 750
    dim = (width, height)
    # resize image
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_LANCZOS4)
    print('Resized Dimensions : ',resized.shape)
    cv2.imwrite(inp_path + filename, resized)
    
print (datetime.datetime.utcnow()) 
```
